```python
#Handling and remove outliers
q1 = payment_df['payment_value'].quantile(.25)
q3 = payment_df['payment_value'].quantile(.75)
iqr = q3-q1

LIF = q1 - 1.5*iqr
UIF = q3 + 1.5*iqr

print('q1:',q1)
print('q3:',q3)
print('iqr:',iqr)
print('LIF:',LIF)
print('UIF:',UIF)
```

```python
#Remove outliers
payment_df = payment_df[(payment_df['payment_value']>=LIF) & (payment_df['payment_value']<=UIF)]
payment_df.reset_index(drop=True, inplace=True)
payment_df
```

###Exploratory Data Analysis (EDA)

####Descriptive Statistics

```python
#Create function for aggregation
def m(x):
    return x.mode()[0]

def q(n):
    def q_(x):
        return x.quantile(n)
    return q_

def selisih(x):
  return x.max()-x.min()

def coeff_var(x):
  return x.std()/x.mean()
```

```python
#Aggregation for descriptive statistics
desc_stat = clean_df['payment_value'].agg(Count=('count'),
                              Min=('min'),
                              Max=('max'),
                              Mean=('mean'),
                              Median=('median'),
                              Mode=m,
                              Q1=q(0.25),
                              Q3=q(0.75),
                              Range=selisih,
                              Variance=('var'),
                              Std=('std'),
                              Coeff_var=coeff_var,
                              Skewness=('skew'),
                              Kurtosis=('kurt')
                              )
desc_stat
```

###Custumer Segmentation

We’re going to determine customer segmentation based on their preference of payment method.  
So, we’re choosing payment_value, payment_type, and payment_installments as the features in this clustering analysis.


```python
#Import library for clustering
from sklearn import cluster
from silhoutte import silhoutte_analysis
from sklearn.preprocessing import MinMaxScaler
```

```python
#Filtering dataframe to remove payment with 0 value
zero_pay = clean_df['payment_value'] == 0
cust_df = clean_df[~(zero_pay)]
cust_df
```

```python
cust_df.info()
```

```python
#Create dummies for payment type
dum_df = pd.get_dummies(cust_df, columns=["payment_type"], prefix=["payment_type"])
df_cluster = clean_df.merge(dum_df)
df_cluster
```

```python
#Normalization of numeric data
scaler = MinMaxScaler()
df_cluster[['payment_installments','payment_value']] = scaler.fit_transform(df_cluster[['payment_installments','payment_value']])
df_cluster
```

```python
df_cluster = df_cluster.drop(['payment_type','order_id','customer_id','order_status','order_purchase_timestamp','order_approved_at','order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date','customer_unique_id','customer_zip_code_prefix','customer_city','customer_state','year-month','weekday'], axis=1)
df_cluster
```

####Elbow method to determine the number of cluster.

```python
distortions = []
K = range(2,11)
for k in K:
    kmeanModel = cluster.KMeans(n_clusters=k)
    kmeanModel.fit(df_cluster)
    distortions.append(kmeanModel.inertia_)

plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

####Silhouette analysis to determine the optimum number of clusters

```python
silhoutte_analysis(df_cluster,list(range(2,5)))
```

Based on the elbow methode & silhouette analysis the most optimum number of clusters is 3.

####Clustering

```python
df = cust_df.copy()
```

```python
#Create cluster model
cluster_model = cluster.KMeans(n_clusters=3)
cluster_model.fit(df_cluster)
cluster_label = cluster_model.labels_
df['cluster'] = cluster_label
df
```

####Cluster Results & Recommendation

```python
#Aggregate cluster data
cluster = df.groupby(['cluster','payment_installments','payment_type'])['payment_value'].agg(['count','mean','median','max','min'])
```
